```python
from ultralytics import YOLO
```

```python
def train_yolo_model():
    model = YOLO("yolo26n.pt")

    yaml_path = '/tf/yolo/yolo_dataset/dataset.yaml'

    project_path = '/tf/yolo/runs'

    print("Starting training...")

    results = model.train(
        data=yaml_path,
        epochs=50,          # עם דאטה גדול, ייתכן שתצטרך יותר מ-30
        imgsz=640,
        batch=64,
        device=0,
        workers=0,          # נסה להעלות (אם קורס, החזר ל-0)
        amp=True,
        cache=True,         # מצוין, מאיץ מאוד אם יש לך מספיק RAM
        # freeze=10,        # הוסר - תן למודל ללמוד הכל
        patience=15,        # תן לו יותר הזדמנות להשתפר
        project=project_path,
        exist_ok=True
    )
    return model
```

```python
model = train_yolo_model()
```

```python
#fine tuning
real_data_yaml = '/tf/yolo/fine_tuning_set/dataset.yaml'
model = YOLO("/tf/yolo/runs/train/weights/best.pt")
results = model.train(
    workers=0,
    device=0,
    data=real_data_yaml,
    epochs=50,             # לרוב מספיק פחות אפוקים מאימון מאפס, תלוי בכמות הדטה האמיתי
    imgsz=640,
    batch=16,              # מומלץ להקטין batch בדאטה אמיתי אם הוא קטן (למשל 16 או 32)
    lr0=0.001,             # קצב למידה התחלתי נמוך יותר (כדי לא "לדרוס" באגרסיביות את מה שנלמד)
    lrf=0.01,              # קצב למידה סופי
    optimizer='auto',      # או 'AdamW' שלפעמים יציב יותר ב-fine-tuning
    project='/tf/runs/fine_tuned',    # נתיב שמירה
    name='finetune_real_clocks', # שם ייחודי לריצה הזו כדי להבדיל מהאימון הסינטטי
    exist_ok=True,
    patience=10,           # סבלנות לעצירה מוקדמת
    save=True,              # שמירת הצ'קפוינטים
    freeze=10              # הקפאת השכבות הראשונות
)
```

```python
import matplotlib.pyplot as plt
for i in range(1,17):

    results = model(f"/tf/yolo/test_image{i}.png")


    # שליפת התוצאה הראשונה (כי הכנסנו תמונה אחת)
    result = results[0]
    annotated_img = result.plot()
    import cv2

    cv2.imwrite(f"/tf/yolo/output_with_boxes{i}.png", annotated_img)
    print(f"Saved to /tf/yolo/output_with_boxes{i}.png")
    # מעבר על כל התיבות שהמודל מצא
    for box in result.boxes:
        # 1. קואורדינטות (בשביל לגזור את התמונה - Crop)
        x1, y1, x2, y2 = box.xyxy[0].tolist() 
        
        # 2. איזה קלאס זה? (0=שעות, 1=דקות)
        class_id = int(box.cls[0])
        label_name = model.names[class_id] 
        
        # 3. כמה המודל בטוח?
        confidence = float(box.conf[0])

        print(f"Found {label_name} at [{x1:.0f}, {y1:.0f}, {x2:.0f}, {y2:.0f}] with {confidence:.2f} confidence")
```
