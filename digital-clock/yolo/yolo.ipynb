```python
from ultralytics import YOLO
```

```python
def train_yolo_model():
    model = YOLO("yolo26n.pt")

    yaml_path = '/tf/yolo/dataset_creator/clock_dataset/dataset.yaml'

    project_path = '/tf/yolo/runs'

    print("Starting training...")

    results = model.train(
        data=yaml_path,
        epochs=50,          # עם דאטה גדול, ייתכן שתצטרך יותר מ-30
        imgsz=640,
        batch=64,
        device=0,
        workers=0,          # נסה להעלות (אם קורס, החזר ל-0)
        amp=True,
        cache=True,         # מצוין, מאיץ מאוד אם יש לך מספיק RAM
        # freeze=10,        # הוסר - תן למודל ללמוד הכל
        patience=15,        # תן לו יותר הזדמנות להשתפר
        project=project_path,
        exist_ok=True
    )
    return model
```

```python
model = train_yolo_model()
```

```python
import matplotlib.pyplot as plt
for i in range(1,18):

    results = model(f"/tf/yolo/test_images/test_image{i}.png")


    # שליפת התוצאה הראשונה (כי הכנסנו תמונה אחת)
    result = results[0]
    annotated_img = result.plot()
    result.plot()
    import cv2

    cv2.imwrite(f"/tf/yolo/output_with_boxes{i}.png", annotated_img)
    print(f"Saved to /tf/yolo/output_with_boxes{i}.png")
    # מעבר על כל התיבות שהמודל מצא
    for box in result.boxes:
        # 1. קואורדינטות (בשביל לגזור את התמונה - Crop)
        x1, y1, x2, y2 = box.xyxy[0].tolist() 
        
        # 2. איזה קלאס זה? (0=שעות, 1=דקות)
        class_id = int(box.cls[0])
        label_name = model.names[class_id] 
        
        # 3. כמה המודל בטוח?
        confidence = float(box.conf[0])

        print(f"Found {label_name} at [{x1:.0f}, {y1:.0f}, {x2:.0f}, {y2:.0f}] with {confidence:.2f} confidence")
```
